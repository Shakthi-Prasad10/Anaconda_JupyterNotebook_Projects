```python
def clean_gender(g):
    g = str(g).strip().lower()
    if g in ['male', 'm', 'man', 'cis male', 'cis man']:
        return 'Male'
    elif g in ['female', 'f', 'woman', 'cis female', 'cis-female/femme']:
        return 'Female'
    else:
        return 'Other'


d['Gender_Clean'] = d['Gender'].apply(clean_gender)
```

```python
d['Gender_Clean'].value_counts()
```

```python
d = d[(d['Age'] >= 18) & (d['Age'] <= 65)]
```

```python
d['Age'].value_counts
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,5))
sns.histplot(d['Age'], bins=10, kde=True)
plt.title("Age Distribution of Respondents")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
d.isnull().sum().sort_values(ascending=False).head(10)
```

```python
d['work_interfere'] = d['work_interfere'].fillna("Don't know")
```

```python
d['benefits'].value_counts(normalize=True) * 100
```

```python
d.groupby('Gender_Clean')['benefits'].value_counts(normalize=True).unstack()
```

```python
plt.figure(figsize=(7,4))
sns.countplot(data=d, x='benefits', order=d['benefits'].value_counts().index)
plt.title("Do Employers Provide Mental Health Benefits?")
plt.xlabel("Benefit Provided")
plt.ylabel("Number of Respondents")
plt.show()
```

```python
benefits_vs_treatment = pd.crosstab(d['benefits'], d['treatment'], normalize='index') * 100
benefits_vs_treatment.round(1)
```

```python
benefits_vs_treatment.plot(kind='bar', stacked=True, figsize=(8,5), colormap='Accent')
plt.title("Treatment Seeking vs. Mental Health Benefits")
plt.xlabel("Benefits Provided")
plt.ylabel("Percentage of Respondents")
plt.legend(title="Sought Treatment")
plt.show()
```
